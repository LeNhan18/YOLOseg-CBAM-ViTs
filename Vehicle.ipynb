```python
# ====================================================
# 🔧 Nếu chưa có yaml, tạo thủ công ở đây
# Điền đúng tên classes của bộ dữ liệu vietnam-vehicle
# ====================================================

# Tự động đọc classes từ label files nếu có
Yaml_path ="/content/drive/MyDrive/yolo_data/vietnam-vehicle/data.yaml"
def get_classes_from_labels(label_dir):
    """Đọc tất cả class IDs từ label files"""
    class_ids = set()
    for lf in glob.glob(f'{label_dir}/*.txt'):
        with open(lf) as f:
            for line in f:
                parts = line.strip().split()
                if parts:
                    class_ids.add(int(parts[0]))
    return sorted(class_ids)

# Mặc định classes cho Vietnam Vehicle (chỉnh nếu cần)
DEFAULT_CLASSES = {
    0: 'motorbike',
    1: 'car',
    2: 'bus',
    3: 'truck',
    4: 'bicycle'
}

# Dùng yaml có sẵn nếu tìm được, hoặc tạo mới
if existing_yamls:
    DATA_YAML = existing_yamls[0]
    print(f" Dùng YAML có sẵn: {DATA_YAML}")
    with open(DATA_YAML) as f:
        print(yaml.safe_load(f))
else:
    # Xác định đường dẫn train/val
    if train_img_dirs:
        train_path = train_img_dirs[0]
        val_path   = val_img_dirs[0] if val_img_dirs else train_img_dirs[0].replace('train', 'val')
        root_path  = str(Path(train_path).parents[1])
    else:
        # Fallback: dataset flat
        root_path  = dataset_root
        train_path = f'{dataset_root}/train/images'
        val_path   = f'{dataset_root}/valid/images'

    data_config = {
        'path':  root_path,
        'train': 'images/train',
        'val':   'images/val',
        'nc':    len(DEFAULT_CLASSES),
        'names': DEFAULT_CLASSES
    }

    DATA_YAML = f'{dataset_root}/data.yaml'
    with open(DATA_YAML, 'w') as f:
        yaml.dump(data_config, f, default_flow_style=False, allow_unicode=True)

    print(f" Đã tạo: {DATA_YAML}")
    print(data_config)

print(f"\n📄 DATA_YAML = '{DATA_YAML}'")
```

```python
import yaml

# Đọc yaml gốc
Yaml_path ="/content/drive/MyDrive/yolo_data/vietnam-vehicle/data.yaml"
EX_DIR = "/content/drive/MyDrive/yolo_data/vietnam-vehicle/"
with open(Yaml_path, 'r') as f:
    cfg = yaml.safe_load(f)

print(" YAML gốc:")
print(cfg)

# ==================================================
# 🔧 Đặt tên class thực tế theo yêu cầu:
# 0: car, 1: motocycle, 2: truck, 3: bus
# ==================================================
CLASS_NAMES = {
    0: 'car',        # class 0
    1: 'motocycle',  # class 1
    2: 'truck',      # class 2
    3: 'bus'         # class 3
}
# ==================================================

# Cập nhật yaml: fix path tuyệt đối + tên class
cfg_fixed = {
    'path':  EX_DIR,
    'train': 'train/images',
    'val':   'valid/images',
    'test':  'test/images',
    'nc':    4,
    'names': CLASS_NAMES
}

FIXED_YAML = f'{EX_DIR}/data_fixed.yaml'
with open(FIXED_YAML, 'w') as f:
    yaml.dump(cfg_fixed, f, default_flow_style=False, allow_unicode=True)

print("\n YAML đã sửa:")
print(cfg_fixed)
print(f"\n Dùng: '{FIXED_YAML}'")
```

```python
import glob, collections
from pathlib import Path

# Đếm số lượng mỗi class trong train set
label_files = glob.glob(f'{EX_DIR}/train/labels/*.txt')
counter = collections.Counter()
for lf in label_files:
    with open(lf) as f:
        for line in f:
            parts = line.strip().split()
            if parts:
                counter[int(parts[0])] += 1

print(" Phân phối class trong Train Set:")
for cls_id, count in sorted(counter.items()):
    cls_name = CLASS_NAMES.get(cls_id, f'class_{cls_id}')
    bar = '█' * (count // 200)
    print(f"  [{cls_id}] {cls_name:12s}: {count:6d}  {bar}")

total = sum(counter.values())
print(f"\n  Total bounding boxes: {total}")
print(f"  Total label files:    {len(label_files)}")

# Cảnh báo mất cân bằng class
if counter:
    max_c = max(counter.values())
    min_c = min(counter.values())
    ratio = max_c / max(min_c, 1)
    if ratio > 5:
        print(f"\n  Mất cân bằng class! Tỉ lệ max/min = {ratio:.1f}x")
        print("   → Sẽ dùng cls_weights trong training để bù.")
    else:
        print(f"\n Class balance OK (ratio = {ratio:.1f}x)")
```

```python
import os, glob, shutil, random
from pathlib import Path

train_img_dir = f"{EXTRACT_DIR}/train/images"
train_lbl_dir = f"{EXTRACT_DIR}/train/labels"

# class repeat factors
# 0:car, 1:motocycle, 2:truck, 3:bus
repeat_map = {
    2: 4,   # truck
    3: 12   # bus
}

img_exts = [".jpg", ".jpeg", ".png", ".webp"]
label_files = glob.glob(f"{train_lbl_dir}/*.txt")

def find_image(stem):
    for ext in img_exts:
        p = Path(train_img_dir) / f"{stem}{ext}"
        if p.exists():
            return p
    return None

created = 0
for lf in label_files:
    stem = Path(lf).stem
    with open(lf, "r", encoding="utf-8") as f:
        lines = [x.strip() for x in f if x.strip()]
    if not lines:
        continue

    classes = set(int(x.split()[0]) for x in lines)
    max_repeat = max([repeat_map.get(c, 1) for c in classes])

    if max_repeat <= 1:
        continue

    src_img = find_image(stem)
    if src_img is None:
        continue

    for r in range(1, max_repeat):
        new_stem = f"{stem}_rep{r}"
        new_lbl = Path(train_lbl_dir) / f"{new_stem}.txt"
        new_img = Path(train_img_dir) / f"{new_stem}{src_img.suffix}"

        shutil.copy2(lf, new_lbl)
        shutil.copy2(src_img, new_img)
        created += 1

print(f"Done. Created {created} oversampled pairs.")
```

```python
# ==================================================
# 🔧 HYPERPARAMETER TỐI ƯU CHO YOLO26
# ==================================================

MODEL_SIZE = 'yolo26m'   # Đổi sang 'yolo26l' nếu có A100

TRAIN_CONFIG = {
    # === Dataset ===
    'data':     FIXED_YAML,
    'imgsz':    640,          # 640 là chuẩn, tăng 1280 cho vật thể nhỏ

    # === Training schedule ===
    'epochs':   150,          # 150 epochs cho 4700 ảnh là hợp lý
    'batch':    16,           # Giảm 8 nếu gặp CUDA out of memory
    'device':   0,

    # === Optimizer – YOLO26 dùng MuSGD mặc định ===
    # Không cần chỉ định optimizer vì YOLO26 tự chọn MuSGD
    'lr0':      0.01,         # LR khởi đầu
    'lrf':      0.01,         # LR cuối = lr0 * lrf
    'momentum': 0.937,
    'weight_decay': 0.0005,
    'warmup_epochs': 3.0,
    'warmup_momentum': 0.8,
    'cos_lr':   True,         # Cosine annealing scheduler

    # === Augmentation mạnh để tăng mAP ===
    'hsv_h':     0.015,
    'hsv_s':     0.7,
    'hsv_v':     0.4,
    'degrees':   5.0,         # Xe thường không nghiêng nhiều
    'translate': 0.1,
    'scale':     0.5,
    'shear':     2.0,
    'perspective': 0.0001,
    'flipud':    0.0,         # Xe không lộn ngược
    'fliplr':    0.5,
    'mosaic':    1.0,         #  Quan trọng – tăng mAP đáng kể
    'mixup':     0.15,
    'copy_paste': 0.1,
    'erasing':   0.4,

    # === Regularization ===
    'label_smoothing': 0.1,
    'dropout':   0.0,

    # === Inference ===
    'conf':      0.001,       # Ngưỡng thấp khi eval để tính đúng mAP
    'iou':       0.7,
    'nms':       False,       # YOLO26 là NMS-free!

    # === Logging & Saving ===
    'patience':  30,
    'save':      True,
    'save_period': 10,
    'cache':     True,        # Cache ảnh vào RAM – train nhanh hơn ~30%
    'workers':   4,
    'plots':     True,
    'val':       True,
    'pretrained': True,       # Luôn bật – transfer learning từ COCO
    'multi_scale': False,     # Bật True nếu muốn detect xe nhỏ & xa
    'overlap_mask': True,

    # === Output – lưu vào GDrive để không mất khi Colab reset ===
    'project':  '/content/drive/MyDrive/yolo26_runs',
    'name':     'vietnam_vehicle_v1',
    'exist_ok': True,
}
# ==================================================

print("  Cấu hình training YOLO26:")
for k, v in TRAIN_CONFIG.items():
    print(f"   {k:20s}: {v}")
```

```python
from ultralytics import YOLO
import time

# Load YOLO26 pretrained trên COCO
model = YOLO(f'{MODEL_SIZE}.pt')
print(f" Loaded: {MODEL_SIZE}.pt  (pretrained on COCO)")
print(f" Bắt đầu training...\n")

t0 = time.time()
results = model.train(**TRAIN_CONFIG)
elapsed = time.time() - t0

print(f"\n{'='*50}")
print(f" Training xong! Thời gian: {elapsed/3600:.2f} giờ")
print(f" Lưu tại: {results.save_dir}")
```

### Debugging và Tiếp tục Training

Nếu training bị dừng đột ngột, có thể là do:
- **Mất kết nối Colab:** Colab có thể tự ngắt kết nối nếu không tương tác trong một thời gian hoặc khi tài nguyên GPU được cấp phát lại.
- **Lỗi bộ nhớ (CUDA out of memory):** Đặc biệt nếu bạn tăng `batch` size hoặc `imgsz` quá cao.
- **Lỗi trong quá trình xử lý dữ liệu:** Có thể có ảnh hoặc label bị lỗi.
- **Patience:** Mặc dù không likely ở epoch 16 với patience là 30, nhưng nếu không có sự cải thiện nào về metric trong 30 epoch liên tiếp, training sẽ dừng sớm.

Bạn nên kiểm tra log output của cell training trước đó để tìm bất kỳ thông báo lỗi nào.

Để tiếp tục training từ checkpoint cuối cùng, bạn có thể sử dụng tùy chọn `resume=True` của Ultralytics. Đầu tiên, hãy kiểm tra xem có những bản chạy nào đã được lưu trong thư mục `project` của bạn.
